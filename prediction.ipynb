```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
word_index=imdb.get_word_index()
reverse_word_index={value:key for key,value in word_index.items()}
```

```python
from tensorflow.keras.models import load_model
model=load_model('model.h5')
model.summary()
```

```python
model.weights
```

```python
#helper function
def decode_review(encode_review):
    return ' '.join([reverse_word_index.get(i-3,'?') for i in encode_review])

def preprocess_text(text):
    words=text.lower().split()
    encodeed_review=[word_index.get(word,2)+3 for word in words]
    padded_review=sequence.pad_sequences([encodeed_review],maxlen=500)
    return padded_review
```

```python
def prediction_sentiment(review):
    review=preprocess_text(review)
    prediction=model.predict(review)
    sentiment='Positive' if prediction[0][0]>0.5 else 'Negative'
    return sentiment,prediction[0][0]
```

```python
prediction_sentiment('This is a very fantastic movie i have seen every till now')
```
